Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.res = 300,repr.plot.width = 5, repr.plot.height = 5)
library('scales')
p= ggplot(penguins, aes(x = body_mass_g, fill = body_mass_g))
p = p +geom_histogram(color = 'blue',fill = 'blue',alpha = 0.5)
p = p +facet_wrap(~year,ncol = 1)
p = p + theme_minimal()
p = p +scale_x_continuous(labels = comma_format())
p = p +scale_y_continuous(labels = number_format())
p = p + theme(
    axis.title.y = element_text(margin = margin(l = 0, r = 10, t = 0, b = 0)),
    axis.title.x = element_text(margin = margin(l = 0, r = 0, t = 10, b = 0)),
    axis.title = element_blank(),
    strip.text = element_text(hjust = 0.93, color = 'grey60', face = 'bold'),
    panel.grid.major = element_line(color = "gray90"),
    panel.background = element_blank()
    )
p = p + labs(title = 'Histograms of penguin body mass for measurements taken from 2007 -2009', x= 'Body mass in grams',y = 'number of penguins')
        
p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
options(repr.plot.res = 300,repr.plot.width = 4, repr.plot.height = 5)
library('scales')
p = ggplot(penguins, aes(y = sex, x = body_mass_g ))
p = p + geom_point(position = 'jitter',color = 'black',fill = 'gray', alpha = 0.5,size = 1)
p = p + facet_wrap(~species,ncol = 1)
p = p + labs(y = '',x= 'Body mass in grams')
p = p + theme_minimal()
p = p +scale_x_continuous(labels = comma_format())
p = p + theme(
    axis.title.y = element_text(margin = margin(l = 0, r = 10, t = 0, b = 0)),
    axis.title.x = element_text(margin = margin(l = 0, r = 0, t = 10, b = 0)),
    strip.text = element_text(hjust = 0.02, color = 'gray40', face = 'bold'),
    panel.grid.major = element_line(color = "gray90"),
    panel.background = element_blank(),
    )       
p
```
